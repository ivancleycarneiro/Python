***Mapa de Calor*** <p>
O termo “mapa de calor” foi originalmente utilizado para identificar áreas com maior ou menor grau de temperatura ambiental. Serve para dimensionar a refrigeração, isolamento térmico e indicar o melhor posicionamento de pessoas e máquinas em um ambiente. Porém, como o termo entrega um entendimento mais fácil dos gráficos de frio e calor, passou a ser utilizado também em outras áreas.

Nesse exemplo usei as bibliotecas Pandas e Plotly para marcar as áreas de calor para representar as vendas relacionadas a cada município, linkando cada um com as coordenadas geográfica.

Nesse caso criei as listas com os dados das cidades, estado e volume de vendas. Depois as cordenadas geográficas das cidades. Quanto maior o  volume de vendas mais avermelhado fica no mapa.

Após criar as listas com os dados, utilizei o Pandas para criar uma Base_Dados em um tabela usando um dicionário(dict) com linhas e colunas preenchendo com as listas criadas no passo anterior.

E para finalizar usando a biblioteca Plotly, configurei a Base_Dados e o tamanho da exibição do mapa. 

Esse foi um exercício proposto como atividade prática de curso de Python para análise de dados.

```python
# Bibliotecas
import pandas as pd
import plotly.express as px
```

```python
# Criar os dados

# Cidades
Cidade = ['Cascavel', 'Foz do Iguaçu', 'Guarapuava', 'Londrina',
          'Maringá', 'Curitiba', 'Pato Branco', 'Guaíra' ]

Estado = ['Paraná', 'Paraná', 'Paraná', 'Paraná',
          'Paraná', 'Paraná', 'Paraná', 'Paraná' ]

# Gerando a Latitude
Latitude = [-24.5721, -25.5469, -25.2326, -23.2927, -23.4273, 
            -25.4284, -23.1346, -24.0794]

# Gerando a Longitude
Longitude = [-53.2719, -54.5882, -51.2715, -51.1732, -51.9375, 
             -49.2733, -52.4018, -54.1527]

# Valor
Vendas = [220, 120, 105, 150, 170, 90, 150, 200]
```

```python
# Dicionario
Dicionario = {
    'Cidade' : Cidade,
    'UF' : Estado,
    'Lat' : Latitude,
    'Log' : Longitude,
    'Vendas' : Vendas
}

Base_Dados = pd.DataFrame( Dicionario )

Base_Dados
```

```python
# Mapa de CAlor Geografico

px.density_mapbox(
    Base_Dados,
    lat='Lat',
    lon='Log',
    z='Vendas',
    radius=30,
    center=dict(lat=-24.5721, lon=-53.2719),
    zoom=5,
    mapbox_style='stamen-terrain'
)
```
